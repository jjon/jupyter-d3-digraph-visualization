# Here we can diddle with data: import graph data as json, and/or generate json data with networkx or rdflib
the json data that d3 expects is as seen in ralphjson below. Directed Multigraph. Can get this from networkx directly.

````python
data = json.load(open("ralph_graph.json"))
ralphjson = open("ralph_graph.json").read()
hostagesjson = open("hostage_graph.json").read()
hos = json.loads(hostagesjson)

## just persons and their links:
persons = [x['id'] for x in hos['nodes'] if 'type' in x and x['type'] == "Person"]
personNodes = [x for x in hos['nodes'] if 'type' in x and x['type'] == "Person"]
personEdges = [e for e in hos['links'] if e['source'] in persons and e['target'] in persons]
persongraph = {'directed': 'true', 'multigraph': 'true', 'graph': {}, 'nodes': personNodes, 'links': personEdges}
pprint(json.dumps(persongraph, indent=4))
````
